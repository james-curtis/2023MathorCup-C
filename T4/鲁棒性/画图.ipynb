```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决无法显示符号的问题
palette = 'deep'
sns.set(font='SimHei', font_scale=1.2, style='whitegrid', palette=palette)  # 解决Seaborn中文显示问题
```

```python
# 生成一个正弦曲线
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# 在y值中添加噪声
y_noise = y + np.random.normal(0, 0.1, size=y.shape)

# 绘制正弦曲线和噪声曲线
plt.plot(x, y, label='原始曲线')
plt.plot(x, y_noise, label='带噪声曲线')

# 添加图例和标签
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

# 显示图像
plt.show()
```
